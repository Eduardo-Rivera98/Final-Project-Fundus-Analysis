```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('full_df.csv')
```

```python
df.labels.value_counts()
```

```python
def has_cataract(text):
    if "cataract" in text:
        return 1
    else:
        return 0
```

```python
df["left_cataract"] = df["Left-Diagnostic Keywords"].apply(lambda x: has_cataract(x))
df["right_cataract"] = df["Right-Diagnostic Keywords"].apply(lambda x: has_cataract(x))
```

```python
df
```

```python
df.drop_duplicates(subset=['ID'],inplace=True)
```

```python
df.drop(['Left-Fundus','Right-Fundus', 'Left-Diagnostic Keywords', 'Right-Diagnostic Keywords', 'filepath', 'target', 'filename'], axis=1, inplace=True)
```

```python
df
```

```python
c=df.groupby(['Patient Age']).agg({'A':'sum'})
```

```python
c
```

```python
c.plot(kind='bar')
```

```python
sns.lineplot(data=c, x="Patient Age", y="A",size=1000)
```
